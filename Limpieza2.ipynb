Aqui continuamos con la limpieza de datos que empezamos en Limpieza1. En Limpieza 2 cambiaremos los datos a valores numericos y normalizaremos el DataFrame

```python
#Importamos las librerias que usaremos
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.preprocessing import normalize 
%matplotlib inline
```

A continuacion vamos a leer el documento csv que se creo al final de Limpieza1

```python
df=pd.read_csv("vgsales_exito.csv")
datanum = []

del df['Unnamed: 0'] #borramos una columna que fue a√±adida erroneamente en los datos
print (list(df))
print(df)
df.head()
```

Ya leido los datos es hora de pasarlos a datos numericos

```python
#Titulos 
dicc_name = dict()
count = 0
array = []
for i in df['Name']:
    if i not in dicc_name.keys():
        count = count + 1
        dicc_name.update({i:count})
    array.append(dicc_name[i])
print(array)
print(dicc_name)
df['Name'] = array
datanum.append(array)
```

```python
#Platforms
dicc_platform = dict()
count = 0
array = []
for i in df['Platform']:
    if i not in dicc_platform.keys():
        count = count + 1
        dicc_platform.update({i:count})
    array.append(dicc_platform[i])
print(array)
print(dicc_platform)
df['Platform'] = array
datanum.append(array)
```

```python
#Year
array = []
a = df['Year'][4]
for i in df['Year']:
    if np.isnan(i):
        i = 0
    array.append(i)
df['Year'] = array
datanum.append(array)
```

```python
#Genre
dicc_Genre = dict()
count = 0
array = []
for i in df['Genre']:
    if i not in dicc_Genre.keys():
        count = count + 1
        dicc_Genre.update({i:count})
    array.append(dicc_Genre[i])
print(array)
print(dicc_Genre)
df['Genre'] = array
datanum.append(array)
```

```python
#Publisher
dicc_publisher = dict()
count = 0
array = []
for i in df['Publisher']:
    if i not in dicc_publisher.keys():
        count = count + 1
        dicc_publisher.update({i:count})
    array.append(dicc_publisher[i])
print(array)
print(dicc_publisher)
df['Publisher'] = array
datanum.append(array)
```

```python
#NA_Sales
array = []
a = df['NA_Sales'][4]
for i in df['NA_Sales']:
    if np.isnan(i):
        i = 0
    array.append(i)
df['NA_Sales'] = array
datanum.append(array)
```

```python
#Exito
exito = []
a = df['Exito'][4]
for i in df['Exito']:
    if np.isnan(i):
        i = 0
    exito.append(i)
#df['Exito'] = array
#datanum.append(array)
```

Ahora normalizamos

```python
columnas = ['Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'Exito']

datanum = normalize(datanum)
datanum = list(datanum)
datanum.append(exito)
arreglo = np.rot90(datanum)
numericDF = pd.DataFrame(data = arreglo,columns = columnas)
#df.loc[df['pad'] == 0,'pad'] = df['pad'].mean()
#numericalDF.head()
#print(array)
numericDF.to_csv('numeric_vgsales.csv')
numericDF.head()
```
