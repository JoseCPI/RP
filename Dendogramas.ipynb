En esta seccion usaremos las herramientas de los dendogramas para ayudarnos a ver el numero de clusters de los datos

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
%matplotlib inline
```

```python

df=pd.read_csv('numeric_vgsales.csv')
del df['Unnamed: 0'] #borramos una columna que fue a√±adida erroneamente en los datos
del df['Exito']
#print (list(df))
#print (df)

x= df.head(20)
df.head()
```

```python
Z = linkage(x, 'single')
plt.figure(figsize=(12, 8))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

En el dendograma anterior usamos el metodo de linkage single. A continuacion usaremos el metodo complete seguido del centroid para ver los distintos agrupamientos que se hacen

```python
Z = linkage(x, 'complete')
plt.figure(figsize=(12, 8))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

```python
Z = linkage(x, 'centroid')
plt.figure(figsize=(12, 8))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

Para los dendogramas anteriores solo consideramos los primeros 20 datos del arreglo. A continuacion usaremos la totalidad de los datos disponibles

```python
Z = linkage(df, 'centroid')
plt.figure(figsize=(12, 5))
dendrogram(Z, leaf_font_size=14)
plt.show()
```

Como se puede ver en el grafico anterior no es posible analizar este grafico en su totalidad, por lo que podemos tomar solo cierta parte del grafico para poder analizarlo

```python
plt.figure(figsize=(12, 5))
dendrogram(
    Z,
    truncate_mode='lastp',  # mostrar s√≥lo los √∫ltims p clusters
    p=10,                   
    show_leaf_counts=True,  # mostrar entre par√©ntesis el n√∫mero de elementos en cada cluster
    leaf_font_size=14.,
)
plt.show()
```

En vista de como termino el dendograma podemos ver como parece que se puede dividir en 3 o tal vez 4 cluster.

```python
#Vamos a ver los ultimos 3 clusters
plt.figure(figsize=(12, 5))
dendrogram(Z, truncate_mode='lastp', p=3, show_leaf_counts=True, leaf_font_size=14.)
plt.show()
```

```python
#Vamos a ver los ultimos 4 clusters
plt.figure(figsize=(12, 5))
dendrogram(Z, truncate_mode='lastp', p=5, show_leaf_counts=True, leaf_font_size=14.)
plt.show()
```
